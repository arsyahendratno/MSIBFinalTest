```python
import requests
import pandas as pd
```

```python
# API URL.
api_url = "https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi"

# 1.
# Making API request.
response = requests.get(api_url)

# Checking whether or not the request is successful.
if response.status_code == 200:
    # Retrieving JSON data from the response.
    json_data = response.json()
    # 2.
    # Extracting value from the data key.
    data_list = json_data.get('data', [])
    # Creating DataFrame from data_list.
    df = pd.DataFrame(data_list)
    # Adding the column 'creator', presuming that the creator is all CNN Indonesia.
    df['creator'] = 'CNN Indonesia'
    # Rename the 'contentSnippet' column to 'description'.
    df.rename(columns={'contentSnippet': 'description'}, inplace=True)
    # Displaying a formatted version of the DataFrame.
    print(df.to_string(index=False))
else:
    # Displaying a message if the request was unsuccessful
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
# Checking the dataframe info.
df.info()
```

```python
# 3.
# Transforming the Dtype of the column 'isoDate' into datetime.
df['isoDate'] = pd.to_datetime(df['isoDate'])
# Checking whether or not the data type has been successfully changed.
df.info()
```

```python
from typing_extensions import Counter

# 4.
# Counting the articles made by a creator.
counter = df.groupby('creator').size().reset_index(name = 'article_count')
# Sorting the data in a descending order.
result = counter.sort_values(by = 'article_count', ascending = False)
# Displaying the result.
print(result)
```
